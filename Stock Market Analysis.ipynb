```python
import pandas as pd
apple = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/AAPL.csv')
```

```python
facebook = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/FB.csv')
```

```python
google = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/GOOGL.csv')
nvidia = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/NVDA.csv')
tesla = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/TSLA.csv')
twitter = pd.read_csv('C:/Personal Stuff/projects/Stock Market Analysis/archive/stocks/TWTR.csv')
```

```python
twitter.head()
```

```python
facebook.head()
```

```python
dfs = [apple, facebook, google, nvidia, tesla, twitter]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
apple.head(200)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
